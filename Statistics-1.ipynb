### Population and Sample

```python
import numpy as np
```

```python
np.random.seed(101)
population = np.random.randint(0,100, size = 100000)
```

```python
population
```

```python
len(population)
```

```python
np.random.seed(101)
sample = np.random.choice(population, 100 )
```

```python
sample
```

```python
population.mean()
```

```python
sample.mean()
```

```python
sample_means = []
for i in range(100):
  sample = np.random.choice(population, 100)
  sample_means.append(sample.mean())

sample_means
```

```python
np.mean(sample_means) 
```

### Dispersion

```python
age = [20, 22, 25, 25, 27, 27, 27, 29, 30, 31]
```

```python
np.std(age)
```

```python
age2 = [20, 22, 25, 25, 27, 27, 27, 29, 30, 31, 121]
```

```python
np.std(age2)
```

```python
np.max(age2) - np.min(age2)
```

```python
np.var(age2)
```

```python
from scipy.stats import iqr
```

```python
iqr(age2)
```

```python
q1 = np.percentile(age2, 25)
q3 = np.percentile(age2, 75)
```

```python
iqr = q3 - q1
iqr
```

```python
q2 = np.percentile(age2, 50)
q2
```

```python
np.median(age2)
```

```python
##skewness
stats.skew(age2)
```

```python
#kurtosiss  .  basıklık degeri
stats.kurtosis(age2)
```
